# After smoting

```python
x= df.copy()
x=x.drop(['Class'], axis=1)
x.head()
```

```python
y= df['Class']
y.head()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
x,y= sm.fit_resample(x,y)
```

```python
df_as= pd.concat([x,y] ,axis=1)
df_as.head()
```

```python
df_as.describe()
```

```python
df_as.info()
```

```python
plt.figure(figsize=(20,16))
sns.countplot(x="Class", data=df_as)
```

```python
corr_n= df.corr()
corr_n
```

```python
from sklearn.utils import shuffle
df_as = shuffle(df_as)
df_as.to_csv('SMOTE_data.csv')
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
Log_model = LogisticRegression(max_iter=500,random_state=5,verbose=1)
```

```python
X= df_as.copy()
X= X.drop(["Class"], axis=1)
X.count()
```

```python
x= X[:341178]
x_t= X[341178:]
```

```python
Y= df_as["Class"]
Y.head()
```

```python
y =Y[:341178]
y_t= Y[341178:]
```

```python
hist = Log_model.fit(x , y)
```

```python
y_p= Log_model.predict(x_t)
print(*y_p)
```

# Analysis of the accuracy of model

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_t, y_p)
```

```python
from sklearn.metrics import f1_score
f1_score(y_t, y_p, average='binary')
```
